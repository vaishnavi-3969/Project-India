```python
import pandas as pd
import numpy as np
data = {
    'Year': ['2017', '2018', '2019','2020'],
    'Total_tb_notifications': [1734965, 2100928,2408724, 893153]
}
df = pd.DataFrame(data)
df
x = df['Year']
y = df['Total_tb_notifications']
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel('Year')
plt.ylabel('Total Tuberculosis notification')
plt.title("Yearwise Total Tuberculosis Notification")
plt.show()
```

```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
y = pd.Series([59, 63, 60, 71, 82, 63])
x = pd.Series([72, 69, 81, 82, 84, 71])
correlation = y.corr(x)
print("Correlation coefficient: ")
print(correlation)
plt.scatter(x, y)
plt.xlabel('Tuberculosis case detection rate (%, all forms)')
plt.ylabel('Tuberculosis treatment success rate (% of new cases)')
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))
         (np.unique(x)), color='red')
plt.title('Correlation graph ')
# plot the data
plt.scatter(x, y)
plt.show()

z = pd.Series([217, 211, 204, 199, 193, 188])
correlation = z.corr(x)
print("Correlation coefficient: ")
print(correlation)
plt.scatter(x, z)
plt.xlabel('Tuberculosis case detection rate (%, all forms)')
plt.ylabel('Incidence of tuberculosis (per 100,000 people)')
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))
         (np.unique(x)), color='green')
plt.title('Correlation graph ')
# plot the data
plt.scatter(x, z)
plt.show()

```

```python
import pandas as pd
import seaborn as sn
import numpy as np
#L1 = Tuberculosis case detection rate (%, all forms)
#L2 = Tuberculosis treatment success rate (% of new cases)
#L3 = Incidence of tuberculosis (per 100,000 people)
data = {    'L1': [72, 69, 81, 82, 84, 71],
            'L2': [59, 63, 60, 71, 82, 63],
            'L3': [217, 211, 204, 199, 193, 188]
        }
df = pd.DataFrame(data)
df
print('''
Pairplot between variables L1, L2, L3
where,
L1 = Tuberculosis case detection rate (%, all forms)
L2 = Tuberculosis treatment success rate (% of new cases)
L3 = Incidence of tuberculosis (per 100,000 people)

''')
sn.pairplot(df)
```

```python
import matplotlib.pyplot as plt

print('\n')
print('Heatmap')
print('\n')
sn.heatmap(df.corr(), annot = True)

sn.set(style="white")
corr = df.corr()
print(corr)



```

```python
def plot_cor_matrix(corr, mask=None):
    f, ax = plt.subplots(figsize=(11, 9))
    sn.heatmap(corr, ax=ax,
                mask=mask,
                # cosmetics
                annot=True, vmin=-1, vmax=1, center=0,
                cmap='coolwarm', linewidths=2, linecolor='black', cbar_kws={'orientation': 'horizontal'})
corr = df.corr()
mask = np.triu(corr)
plot_cor_matrix(corr,mask)
plt.show()
```
